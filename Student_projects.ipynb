```python
### JAKE: I don't think you were competing but I decided to showcase your function anyway.
### Great piece of code!

#
# fibonaccy.py
#
# Given a number R, writes the N-th Fibonacci number
#
# If R is a natural number N:
#   N = 0, Fibonacci(N) = 1
#   N = 1, Fibonacci(N) = 1                                       [1]
#   N > 1, Fibonacci(N) = Fibonacci(N-1) + Fibonacci(N-2)
#          i.e., Fibonacci(2) = 2, Fibonacci(3) = 3, Fibonacci(4) = 5
#
# If R is a real number R:
#   phi = (1+sqrt(5))/2,
#   Fibonacci(R) = (phi**(R+1) - (-phi)**(-R-1))/sqrt(5)                [2]
#
# If R is a natural number Fibonacci(R) produces the same result as Fibonacci(N)
#
# example:
#  from fibonacci import Fibonacci
#  print(Fibonacci(6))
#
# (c) Alle Meije Wink 2017
#

# function definition
def Fibonacci ( input = 0 ):

    # for nonnegative integers, compute iteratively
    if isinstance( input, int ) and input >= 0:

        # list elements for N = 0 and N = 1
        fib = [ 1, 1 ];

        # add new elements to the list for N > 1
        for x in range( 1, input ):
            fib.append (fib [ -1 ] + fib [ -2 ])

        # the answer is at the end of the list
        return fib [ -1 ]

    # for nonnegative floats, compute analytically
    elif isinstance( input, float ) and input >= 0:

        # golden section phi
        phi = ( 1 + 5**(.5) ) / 2
        # R + 1
        in2 = input + 1

        # use the formula above
        return ( ( phi ** in2 - (-phi) ** (-in2) ) / 5**(.5) )

    else:

        # for non-numbers and negative values there is no answer (non-integers produce complex answers)
        raise ValueError ( "Fibonacci number can only be conputed for nonegative numbers" )
```
